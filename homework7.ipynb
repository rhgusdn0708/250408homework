```python
def classify_gender_ratio(ratio):
    if ratio > 1.03:  # 남여 비율이 1.03보다 크면 남초
        return "남초 지역"
    elif ratio < 0.97:  # 남여 비율이 0.97보다 작으면 여초
        return "여초 지역"
    else:  # 그 사이는 균형 지역
        return "균형 지역"

# '지역 유형' 컬럼 추가
result['지역 유형'] = result['남여 비율'].apply(classify_gender_ratio)

# 결과 출력
print(result.sort_values(by='남여 비율', ascending=False))

# 지역 유형별 개수 확인
print("\n지역 유형별 개수:")
print(result['지역 유형'].value_counts())
```

```python
# '전국' 제외한 데이터 준비
filtered_data = data[data['행정기관'] != '전국'].copy()

# 세대당 인구 평균 계산
average_household_pop = filtered_data['세대당 인구'].mean()

# 평균보다 높은 지역 필터링
above_avg = filtered_data[filtered_data['세대당 인구'] > average_household_pop]

# 결과 정렬 및 컬럼 선택
result = above_avg[['행정기관', '세대당 인구']].sort_values(by='세대당 인구', ascending=False)

# 평균값 출력
print(f"전국 세대당 인구 평균: {average_household_pop:.2f}\n")

# 결과 출력
print("세대당 인구가 평균보다 높은 지역:")
print(result.reset_index(drop=True))
```
